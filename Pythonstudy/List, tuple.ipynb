```python
# 리스트 연산하기
li4 = [1,2,3]
li5 = [3,4,5]

print(li4*3)
print(li4+li5)
#print(li4*li5)
```

```python
#리스트 값 추가하기
# 리스트.append(값)
li6=[1,4,3,7,6]
li6.append(5)
print(li6)
```

```python
#리스트 값 삽입
#리스트명,insert()

li7= [0,1,2,3,4]
li7.insert(1,5)
print(li7)
```

```python
#리스트 값 수정
# 리스트명[인덱스]=값
# 리스트명[슬라이싱]= [값]
li8=[0,1,2,3.4]
li8[1]=7
print(li8)
```

```python
array=[1,2,3,4,5]
# 숫자 5를 7로 변경
array[-1]=7
print(array)
array[1]=['a','b','c']
print(array)
array[3:]='d','e','f','g'
print(array)
array[:1]='h','i','j'
print(array)
```

```python
# 리스트 값 삭제
# 인덱스 이용하는 방법
# -del 리스트명[인덱스]
# 2. 값을 이용하는 방법
# -리스트명 .remove(값)
li9=['c','a','b','c']
del(li9[0])
li9.remove('c')
print(li9)
```

```python
#리스트 정렬
li10=[9,77,13,51,100,3]
li10.sort()
print(li10)
#리스트 순서 뒤집기
li10.reverse()
print(li10)
```

```python
# 리스트 순서 뒤집기
# 리스트명.reverse()
# 함수정보 확인 단축키 shift + Tab
li11=[9,77,13,51,17,100]
li11.sort(reverse=True)
print(li11)
```

```python
str1= '파이썬 최고'
print('파이썬' in str1)
print('파이썬' not in str1)

li=[77, 38, 10]
print(33 in li)
print(33 not in li)

if '파이썬' in str1:
    print("포함")
else:
    print("미포함")
    
```

```python
lst=['딸기','바나나', '수박','체리','포도']
print('*'*50)
char=input('검색할문자를 입력하세요>> ')
if char in lst:
    print(f'{char}은/는 리스트에 {lst.index(char)}번째 인덱스에 들어있습니다.')
else:
    print(f'{char}은/는 리스트에 들어있지 않습니다.')
print('*'*50)
```

```python
sentence= "Hi, My name is Mihee"
print('*'*50)
char=input("검색할 문자를 입력하세요: ")
if char not in sentence:
    print(f'{char}는 문자열에 들어있지 않습니다.')
else:
    print(f'{char}는 {sentence.count(char)}번 들어있습니다.')
print('*'*50)    
```
