```python
import heapq
#    
my_heap = []
N = int(input())
for i in range(N):
    num = int(input())
    if num == 0:
        if not my_heap:
            print(0)
        else:
            a = heapq.heappop(my_heap)
            print(a)
    else:
        heapq.heappush(my_heap, num)
 # 1        
```

```python
import heapq
import sys
N = int(sys.stdin.readline().rstrip()) 
my_heap = []
for i in range(N):
    num = int(sys.stdin.readline().rstrip())
    if num != 0:
        heapq.heappush(my_heap, num)
    else:
        if my_heap:
            print(heapq.heappop(my_heap))
        else:
            print(0)
# 2  sys  0   
```
