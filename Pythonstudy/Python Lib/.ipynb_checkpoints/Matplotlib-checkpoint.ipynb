# matplotlib.pyplot 임포트하기

```python
import matplotlib.pyplot as plt
```

# plt.plot

```python
y = [2, 4 ,6]

# plot -선 그래프(Line plot)
plt.plot(y)
plt.show
```

```python
y = [2, 4, 6]
x = [1, 2, 3]
# plot -선 그래프(line plot)
plt.plot(x,y)
plt.show
```

## plt : line style

```python
#line style --> ls로도 바꿀수 잇음
plt.plot(x,y ,linestyle = '--')
plt.show
```

## plt : marker 설정

```python
# Marker 
plt.plot(x,y ,marker ='o')
plt.show
```

```python
x = [2, 4, 6, 8]
y1 = [10, 13, 16, 20]
y2 = [10, 20, 15, 12]

plt.plot(x, y1 , ls ='-', marker = 'o')
plt.plot(x, y2 , ls =':')
plt.show
```

## line color, line width

```python
# line width 설정
plt.plot(x, y1 , linewidth=100, color = 'violet')
```

```python
#line color 설정
plt.plot(x, y1, color = 'cyan')
```

```python
#marker color 마커 색상
plt.plot(x,y1, color = '#00ff66' , marker ='D' , mfc ='#7f7f7f')
plt.show
```

## 연습문제

```python
import numpy as np
x =np.arange(7)
y=[1,4,5,8,9,5,3]
```

```python
plt.plot(x,y, ls ='-.' ,color = '#A8DF8E', marker ='o' ,
         mfc = '#0C356A' , linewidth = 3 , mec = '#CD6688', mew =5 ,ms=15 )
plt.grid()
plt.xticks([0,3,6]) #x축 틱 설정
plt.yticks([1,2,3]) #y축 틱 설정
plt.show()
```

```python
# 여러 plot 그리기
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [3, 6, 9, 12]
# 그래프에 이름 부여
plt.plot(x , label ='x')
plt.plot(y, label ='y')
plt.plot(z, label ='z')
plt.legend() #범례
plt.show()
```

```python
# 한글 지원 폰트 변경
from matplotlib import rc
rc('font', family = 'Malgun Gothic')
```

```python
plt.plot(x , label ='엑스')
plt.plot(y, label ='와이')
plt.plot(z, label ='지')
# 축에 이름 표시
plt.xlabel('x축')
plt.ylabel('y축', rotation = 0)
# 그래프 이름
plt.title('x,y,z 그래프')
plt.legend()
plt.show()
```
