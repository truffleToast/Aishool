```python
ratings = np.loadtxt("ratings.txt", delimiter = ',' , dtype = np.int64)
ratings
```

```python
print(ratings.ndim)
print(ratings.shape)
print(ratings.size)
```

```python
# 전체 평점의 평균 구하기
print(ratings[:,2].mean())
print(np.mean(ratings[:,2]))
```

```python
# 유저 아이디가 1인 유저의 평점 평균 구하기
print(ratings[ratings[:,0]==1,2].mean())
```

```python
print(ratings[ratings[:,0]==5,2].mean())
```

```python
# 고유한 요소 1개씩 남겨서 배열 생성
unique_id=np.unique(ratings[:, 0])
li=[]
for user_id in unique_id :
    b= ratings[:,0] == user_id
    user_mean = ratings[b,2].mean()
    li.append([user_id, user_mean])
```

```python
li[:5]
```

```python
# 영화별 평점 구하기
unique_movies = np.unique(ratings[:,1])
movies=[]
for movie in unique_movies :
    m = ratings[:,1] == movie
    movie_mean = ratings[m,2].mean()
    movies.append([movie, movie_mean])
```
