### 스코어 실습

```python
score =pd.read_csv("score.csv", encoding = 'cp949', index_col='과목')
score
```

```python
# axis = 0 일때는 반별 합계(열 기준으로)
score.sum()
```

```python
# axis =1 일때는 과목별 합계(행 기준)
score.sum(axis=1).sort_values(ascending = False) 
```

```python
# 학급별 총합
score.sum(axis = 0).sort_values(ascending = False)
```

```python
# 합계 컬럼 생성
#1. 과목별 총합 값 구하기
#2. 컬럼 생성
```

```python
hap=score.sum(axis=1)
```

```python
score['합계'] = hap
```

```python
score
```

```python
#2 . 컬럼생성
score['합계'] =score.loc[: ,'1반':'4반'].sum(axis =1)
```

```python
score
```

```python
score['평균'] =score.loc[:,'1반':'4반'].mean(axis=1)
score
```

```python
# 반평균 데이터 생성
#1. 반평균 값 구하기
#2.  1을 이용해 반평균 삽입
classmean= score.loc[:,'1반':'평균'].mean(axis=0)
classmean
```

```python
score.loc['반평균']=classmean
```

```python
score
```

```python
# 과목별 최대값과 최소값의 차이
```

```python
max_score =score.max(axis=1)
min_score =score.min(axis=1)
```

```python
score.loc[:'Web', '1반':'4반']
```

```python
mas=score.loc[:'Web', '1반':'4반'].max(axis=1)
mas
```

```python
mis=score.loc[:'Web', '1반':'4반'].min(axis=1)
mis
```

```python
mas-mis
```

```python
#apply 함수 적용
# 행 혹은 열단위로 복잡한 데이터 프레임을 처리할 때 사용
```

```python
# 함수 정의
# 과목별 max 와 min 점수 차이 구하기
def max_min(x):
    return x.max()-x.min()
```

```python
score.loc[:'Web', '1반':'4반'].apply(max_min, axis =1)
```

### 카테고리 생성하기
* 데이터를 구간별로 구분하기!

```python
ages = [0,2,10,35,6,112,1,50,9,10,18,1,521,5,23,60,40,1,251,1,21,3,4,5,2,61,60,2,12,80,90,91]
bins = [0,5,25,35,50,100,250, 600] # 구간 labels 변수에 넣기
labels= ['미취학아동','학생','백수', '장년', '노인', '장수','미라'] # 구간의 이름
# 1~5 까지 미취학 아동, 6~ 24까지 학생 26~35 : 백수 36~ 50 까지 장년, 51~100 까지 노인 , 101~250 장수 , 251~ 미라
cats = pd.cut(ages, bins, labels =labels)
```

```python
cats
```

```python
cats.value_counts
```

```python
age =pd.DataFrame(ages, columns =['나이'])
age
```

```python
age['연령대']=cats
```

# 데이터 프레임 병합
* merge 
* concat

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'a':['a0','a1','a2','a3'],
                   'b':['b0','b1','b2','b3'],
                   'c':['c0','c1','c2','c3']},
                  index = [0,1,2,3])

df2 = pd.DataFrame({'a':['a2','a3','a4','a5'],
                   'b':['b2','b3','b4','b5'],
                   'c':['c2','c3','c4','c5'],
                   'd':['d2','d3','d4','d5']},
                   index = [2,3,4,5])

df3 = pd.DataFrame({'a':['a3','a4','a5','a6'],
                   'b':['b3','b4','b5','b6'],
                   'c':['c3','c4','c5','c6'],
                   'd':['d3','d4','d5','d6']},
                   index = [3,4,5,6])
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
# DataFrame 간 병합
resultDf1 = pd.concat([df1,df2])
resultDf1
```

```python
# 기존 인덱스를 무시하고 새로운 인덱스를 부여
resultDf2 = pd.concat([df1,df2,df3], ignore_index=True)
resultDf2
```

```python
#행(인덱스 라벨)을 기준으로 병합
resultDf3 =pd.concat([df1,df2], axis =1)
resultDf3
```

```python
# inner join
resultDf4 = pd.concat([df1,df2,df3], axis=1, join='inner')
resultDf4
```

```python
#Series간 병합
sr1 = pd.Series(['e0','e1','e2','e3'], name = 'e')
sr2 = pd.Series(['f0','f1','f2'], name = 'f')
sr3 = pd.Series(['g0','g1','g2','g3'], name = 'g')
```

```python
resultDf5 =pd.concat([sr1,sr2], axis=1)
resultDf5
# 일치하지 않는 데이터에는 nan이 출력.
```

```python
#Series와 DataFrame 병합
sr4 = pd.Series(['h0','h1','h2','h3'], name = 'h', index = [1,3,4,5])
sr4
```

```python
resultDf6 =pd.concat([df2,sr4], axis=1)
resultDf6
```

```python
dic1 = {'key' : ['K0','K2','K3','K4'], 'A' : ['A0','A1','A2','A3'], 'B' : ['B0','B1','B2','B3']}
df4 = pd.DataFrame(dic1)
```

```python
dic2 = {'key' : ['K0','K1','K2','K3'], 'C' : ['C0','C1','C2','C3'], 'D' : ['D0','D1','D2','D3']}
df5 = pd.DataFrame(dic2)
```
