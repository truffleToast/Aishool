```python
while True:
    num1=int(input("첫번째 정수 입력 >>"))
    num2=int(input("두번째 정수 입력 >>"))
    if num1 == 0 and num2 == 0:
        print("프로그램이 종료되었습니다.")
        break
    print(f'두 정수의 합: {num1+num2}')
```

```python
# 단 음수를 허용하지 않아야 가능 바람직한 것은 or 
num1, num2 = 0 ,1
while num1 ! == 0 or num2 ! == 0
    print(f"두 정수의 합: {num1+num2}")
    num1=int(input("첫번째 정수 입력 >> "))
    num2=int(input("두번째 점수 입력 >> "))
    if num1 ! == 0 or num2 ! == 0:
        print(f'두수의 합: {num1+num2}')
print('프로그램 종료')        
```

```python
import pickle  #pickle모듈을 이용해 데이터를 로드해보자
    with open('data/movie_review.pkl', 'rb') as f:
        mr= pickle.load(f)
```

```python
print(len(mr))
```

```python
smr=0  
for m in mr:
    smr+=len(m[0])
am= smr/len(mr)
print(f'리뷰의 평균 길이는 "{am}"입니다.') 
```

```python
####긍정리뷰/ 부정리뷰
pos, neg=0, 0
for m in mr:
    if m[1]==1:
        pos +=1
    else:    
        neg +=1
print(f"긍정 '{pos}'개\n부정 '{neg}'개")
```

```python
## 리뷰길이가 가장 긴 거 찾기
review_length= len(mr[0][0])
maxtxt=0
for txt,pn in mr:
    if review_length <len(txt):
        review_length =len(txt)
        maxtxt=txt
        
print(f'리뷰 최대길이는 {review_length}입니다')
print(maxtxt)
        
```

```python
### 검색 데이터가 포함된 개수
data= input("검색할 데이터 입력>> ")
creview=0
for txt, pn in mr:
    if data in txt:
        creview+=1
print(f"{data}'가 포함된 리뷰는 {creview}개 입니다.")        
```
