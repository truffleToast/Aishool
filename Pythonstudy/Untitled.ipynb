#heap은 tree구조 처럼 생겼다. 


maxheap은 가장큰 값을 루트 노드로하고 minheap은 가장 작은 값을 루트 노드를 하는 특징을 가졌다.
파이썬에서는 최소힙만 제공하기에 알고리즘적으로 최대힙을 구현해야 한다,

```python
import heapq
# 첫째줄에 연산의 개수를 부여
my_heap = []
N = int(input())
for i in range(N):
    num = int(input())
    if num == 0:
        if not my_heap:
            print(0)
        else:
            a = heapq.heappop(my_heap)
            print(a)
    else:
        heapq.heappush(my_heap, num)
 # 1차시기 시간초과       
```

```python
import heapq
import sys
N = int(sys.stdin.readline().rstrip()) 
my_heap = []
for i in range(N):
    num = int(sys.stdin.readline().rstrip())
    if num != 0:
        heapq.heappush(my_heap, num)
    else:
        if my_heap:
            print(heapq.heappop(my_heap))
        else:
            print(0)
# 2차 시기 sys로 바꾸고 0이 아닌것을 먼저 처리
```

```python
# 최대힙 문제 파이썬에는 그런데 최대힙이라는게 존재하지 않는다. 그러므로 다른 접근이 필요하다.
# 가장 높은 수를 (-1)로 곱하면 최소힙이 완성되고 거기에 (-1)을 곱하면 유사 최대힙이 될 수 있다.
import heapq
import sys
N = int(sys.stdin.readline().rstrip()) 
my_heap = []
for i in range(N):
    num = int(sys.stdin.readline().rstrip()) *(-1)
    if num != 0:
        heapq.heappush(my_heap, num)
    else:
        if my_heap:
            print(heapq.heappop(my_heap)*(-1))
        else:
            print(0)
# 2차 시기 sys로 바꾸고 0이 아닌것을 먼저 처리
```
