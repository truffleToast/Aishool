```python
def number_sum(num1, num2):
    r= num1+num2
    return r
```

```python
def cal(num1, num2=1 ,op='+'): # 매개변수 기본값 호출하기
    """덧셈과 뺄셈을 하는 함수"""
    if op == '+':
        return num1+num2
    else:
        return abs(num1-num2)
```

```python
num1= int(input("첫번째 정수입력>> "))
num2= int(input("두번째 정수입력>> "))
op= input("연산자 입력(+,-)>> ")
result= cal(num1, num2, op)
print("결과: {}".format(result))
```

```python
cal(3,4,'-')
```

```python
print(cal(7,op='-'))
print(cal(8))

print(cal('A','B'))
```

# 가변 메소드, 매개변수


```python
def add(*args):
    answer= 0
    for nums in args:
        answer+=nums
    return answer
```

```python
add(1,2,3)
add(4,5,6,7,8,9,10)
```

```python
print(add(1,2,3))
print(add(4,5,6,7,8,9,10))
```
