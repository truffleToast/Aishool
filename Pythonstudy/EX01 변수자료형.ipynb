## 인덱싱과 슬라이싱
#### ● 인덱싱 - 무언가를 '가리킨다'는 의미

### 슬라이싱 
#### ● 무엇인가를 '잘라낸다'는 의미


```python
s= "My name is MH"
```

```python
#인덱싱
print(s[3])
print(s[-1])
```

```python
#슬라이싱
# 변수명[0~1] ->  변수명 [0:2]
## 시작 인덱스가 0이거나 끝나는 경우(-1)일 때 생략할 수 있다.
print(s[0:2])
print(s[:7])
print(s[8:])
```

```python
#문자열 안에 특정 값을 바꿔야 할 경우
month= 7
day=3
# 문자열과 숫자열 데이터는 더하기 연산이 불가
# s= '오늘은' + month+ '월 ' +day+ '일입니다.'
#print(s)
```

```python
# 문자열 s만들기
s='오늘은 %d월 %d일 입니다.' %(month, day)
print(s)
```

```python
# format 함수를 이용한 포매팅
s= '오늘은 {}월 {}일 입니다.'.format(month, day)
print(s)
```

```python
# f문자열 
s= f"오늘은 {month}월 {day}일 입니다."
print(s)
```

```python
x=100
y=200
sum2=x+y
s1='{}와 {}의 합은 {}입니다.'.format(x,y,sum2)
print(s1)
print('%d와 %d의 합은 %d입니다.'%(x, y, sum2))
s2=f'{x}와 {y}의 합은 {sum2}입니다.'
print(s2)
```

## 문자열 관련함수


```python
##문다 개수 세기
s='My name is MH'
print(s.count('i'))
```

```python
# 위치 알려주기
#문자열.find(데이터): 찾는 문자가 존재하지 않을 경우 -1을 반환
# 문자열.index(데이터): 찾는 문자가 존재하지 않을 경우 에러
print(s.find('z'))
print(s.index('e'))
```
