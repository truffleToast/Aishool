```python
time =int(input("시간 입력(초)"))
hour= time // 3600
minute = time % 3600 // 60
second =  time % 3600 % 60
print('{}시간 {}분 {}초'.format(hour, minute, second))
print(f'{hour}시간 {minute}분 {second}초')
print('%d시간 %d분 %d초'%(hour, minute, second))
```

- 논리연산자
   * 자바 : && || !
   * python : and or not

```python
a=3
b=7
print(not a<b)
print(a>b and 10 == 20)
print(a<b and 10 != 20)
print(a<b or 10==20)
print(True or False)
```

* 논리 연산자:
    

```python
# a if 조건식 else b
score=80
result = '합격' if score>=60 else '불합격'
print(result)
```

```python
a=int(input('정수입력>> '))
result = '짝수' if a % 2 ==0 else '홀수'
print(f'{a}는 {result}입니다')
```

```python
if True :
    print('실행문1')
    print('실행문2')
    if True:
        print('실행문3')
print('조건문 종료')    
```

```python
score=int(input("점수입력:"))
if score>=90:
    print(f'{score}는 A학점입니다.')
elif score>=80:
    print(f'{score}는 B학점입니다.')
elif score>=70:
    print(f'{score}는 C학점입니다.')
elif score>=60:
    print(f'{score}는 D학점입니다.')
else:
    print(f'{score}는 F학점입니다.')     
    
```

```python
print('--------[SM COFFEE]-------------')
money = int(input('금액을 입력하세요 >>'))
print('커피를 고르세요.\n[1]에스프레소(600원) [2] 아메리카노(1000원) [3]카페라떼(1300원) >>' ,end='')
menu=int(input())       
if menu==1:
    coffee_price=600
elif menu==2:
    coffee_price=1000
elif menu==3:
    coffee_price=1300
else:
    print("잘못입력했습니다.")
if money>=coffee_price:
    print('거스름돈은 {}입니다.'.format(money-coffee_price))
else:
    print("돈이 부족해요\n거스름돈은 {}입니다".format(money))
    
```
